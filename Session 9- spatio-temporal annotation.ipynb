#### STEP 2: Find out referent tile and percentage of gaze points based on the ranking algorithm

This is the outcome of the last notebook we used (session 8)

```python
tolerance  = 75
results = []
for episodeno in infos['episodes'].index:
    if episodeno in [36, 53, 142]:
        continue
    aboard = pd.DataFrame(get_board(episodeno, infos, boards))
    aboard['tile'] =  aboard[aboard.columns[0]]
    del aboard[aboard.columns[0]]
    aboard['looked_at'] = aboard['tile'].map(lambda x: get_number_of_gaze_points(get_gaze(episodeno, infos, gazegood), 
                                                                                 x, tolerance))
    aboard['selected'] = aboard['tile'].map(lambda x: x.is_selected)
    aboard.sort(columns='looked_at', ascending=False, inplace=True)
    aboard.reset_index(inplace=True)
    aboard['rank'] =aboard.index + 1
    total = aboard['looked_at'].sum()
    aboard['percentage'] = aboard['looked_at'].map(lambda x: round(100. * x / total, 2))
    rank = list(aboard['rank'][aboard['selected']])[0] #item 0 of the list is the one item in the list
    percentage = list(aboard['percentage'][aboard['selected']])[0]
    result = {'episodeno': episodeno, 'rank': rank, 'percentage': percentage}
    results.append(result)
RESULTS = pd.DataFrame(results)
RESULTS
```

#### STEP 3: Derive an interval frame with the tiles looked at for each episode

At the moment we have only information about *where* people look, but not *when*. E.g. it would be good to know the succession of looking at tile objects, e.g. first at tile 1, then at tile 15 etc.

We need to devise an algorithm to yield *which* object is being looked at at any one time.

* find out which tiles have more than zero points on them
* for each tile, find all gaze points near it
* create an intervalframe for each tile separately
* Compute the union of all intervalframes in order to get the final result

We go through this step by step below

```python
#3.1 Find all tiles with non-zero points
episodeno = 40
aboard = pd.DataFrame(get_board(episodeno, infos, boards))
aboard['tile'] =  aboard[aboard.columns[0]]
del aboard[aboard.columns[0]]
aboard['looked_at'] = aboard['tile'].map(lambda x: get_number_of_gaze_points(get_gaze(episodeno, infos, gazegood), 
                                                                                 x, tolerance))
lookedat = aboard[aboard['looked_at'] > 0] 
```

```python
aboard
```

```python
lookedat
```

```python
#3.2 find all gaze points near each of the tiles above
episode_gaze = get_gaze(episodeno, infos, gazegood)
for tindex in lookedat.index:
    t = lookedat['tile'].ix[tindex]
    episode_gaze[tindex] = episode_gaze['pixels'].map(lambda x: distance.euclidean((x.x, x.y), (t.x, t.y)) < tolerance)
```

```python
episode_gaze;
```

```python
#3.3 Creare an intervalframe for each tile separately
tilename = 'object2'
t2 = create_intervalframe_from_streamframe(episode_gaze, tilename, 
                                           lambda x: True if x else False, 
                                           200, text = tilename)
tilename = 'object9'
t9 = create_intervalframe_from_streamframe(episode_gaze, tilename, 
                                           lambda x: True if x else False, 
                                           200, text = tilename)
tilename = 'object6'
t6 = create_intervalframe_from_streamframe(episode_gaze, tilename, 
                                           lambda x: True if x else False, 
                                           200, text = tilename)
t2['start_time'] /= 1000.
t2['end_time'] /= 1000.
t9['start_time'] /= 1000.
t9['end_time'] /= 1000.
t6['start_time'] /= 1000.
t6['end_time'] /= 1000.
```

```python
plot_annotations({'tile 2': t2, 'tile 9': t9, 'tile 6': t6}, time_begin=980, scalefactor=1)
```

```python
#3.4 Compute the uninon
tjoined = t2.append(t9).append(t6)
tjoined.sort(columns='start_time', inplace=True)
tjoined.reset_index(inplace=True)
del tjoined['index']
```
