### Export Textgrids and Plots to ELAN and Praat

```python
%matplotlib inline
import pandas as pd  
from mumodo.mumodoIO import open_streamframe_from_xiofile, save_intervalframe_to_textgrid, convert_streamframe_to_pointtier
from mumodo.analysis import plot_scalar, create_intervalframe_from_streamframe
```

#### Import head-tracking data

```python
xiofilename='../../DATA/sampledata/r1_12_15.xio.gz'
```

```python
facelab = open_streamframe_from_xiofile(xiofilename,'lab-labtop/FaceLabNode')
#facelab.index -= facelab.index[0]
```

Just an idea of what kind of data we have just imported

```python
facelab.columns
```

#### How well was the head tracked?

We can plot the confidence of head tracking for the whole file or just a specified time

```python
#plot_scalar(facelab, ['headPositionConfidence'])
plot_scalar(facelab, ['headPositionConfidence'], 0, 1000) #times are in milliseconds !
```

But we don't want to look at data with low confidence, so we want to set a threshold below which we ignore data!

```python
threshold = 0.4
facelab['threshold'] = facelab['time'].map(lambda x: threshold)
```

We can look at how much data we drop by looking at the plot again, only this time we plot our threshold, too

```python
plot_scalar(facelab, ['headPositionConfidence', 'threshold'])
```

#### Export data to formats readable by other programs

First of all, we can create a Praat Textgrid with the intervals in which we deem the confidence satisfactory

```python
goodintervals = create_intervalframe_from_streamframe(facelab, 'headPositionConfidence', lambda x: True if x>threshold else False, 500)
```

```python
goodintervals
```

Praat prefers seconds to milliseconds. In addition, we change our label that will appear in Praat/ELAN on each interval

```python
goodintervals['start_time'] /= 1000. 
goodintervals['end_time'] /= 1000. 
goodintervals['text'] = goodintervals['text'].map(lambda x: " > " + str(threshold))
goodintervals                                   
```

We now save our data in the Textgrid format

```python
save_intervalframe_to_textgrid({'confidence': goodintervals}, "testgrid.TextGrid")
```

In addition, we can save our data also as a series in standard CSV format

```python
facelab.dropna(subset = ['headPositionConfidence'], inplace=True)
```
