#### STEP 3: Define the plotting functions

```python
#Function that plots gaze points and objects of the board
def plot_screen(tiles, gazepoints, tolerance=0, filepath=None):
    """ plot an episode screen
    
    tiles -- a list of tile objects
    gaze points -- a series of sfvec2f objects
    
    filepath -- optional argument to save the plot to
                a file. The filepath MUST be have a 
                valid image extension, e.g. "file.png"
                (bmp, jpg, etc)
    
    """
    
    #read gaze points
    gazeX = gazepoints['pixels'].map(lambda x: x.x)
    gazeY = gazepoints['pixels'].map(lambda x: x.y)
    gazeI = gazepoints.index
    
    
    #create figure
    fig = plt.figure(figsize=(16,9))
    ax = fig.add_subplot(111)
    
    ax.set_xlim(-200,2120)
    ax.set_ylim(-200,1280)
    ax.set_ylim(ax.get_ylim()[::-1])
    ax.add_patch(plt.Rectangle((0,0),1920,1080,alpha = 0.05,color = 'b'))
    gax = ax.scatter(gazeX, gazeY, s= 10, c = gazeI, cmap = plt.cm.Reds, label = 'gaze')
    fig.colorbar(gax, format = '%d')
    
    #draw the objects
    tolerances = []
    for tile in tiles:
        if tile.is_selected:
            ax.text(tile.x, tile.y, tile.shape, size = 'xx-large', weight = 'bold', 
                        color = tile.color, bbox=dict(facecolor='red', alpha=0.5))   
        else:
            ax.text(tile.x, tile.y, tile.shape, size = 'xx-large', weight = 'bold', color = tile.color)
        #draw the tolerance circles
        circle = plt.Circle((tile.x, tile.y), tolerance, color='r', fill=False)
        ax.add_patch(circle)
    if filepath is not None:
        fig.savefig(filepath, format=filepath.split('.')[1])
        
```

```python
#function that plots an episode based on its number
def plot_episode(episodeno, tolerance=0, filepath=None):
    episode_start = int(infos['episodes']['start_time'].ix[episodeno])
    episode_end = int(infos['episodes']['end_time'].ix[episodeno])
    gazeslice = gazegood.ix[episode_start:episode_end]
    boardslice = boards.ix[episode_start:episode_end]
    plot_screen(boardslice.iloc[-1], gazeslice, tolerance, filepath)
```

### INTERMISSION: A Research Question

When doing research, we look (and process) data in order to answer *questions*. In linguistics and cognitive science, we may be interested in modeling (i.e. arriving at some knowledge of) the behaviour of (human) agents in communicative situations.

Here we have a communicative task. The human participant observes the "world" (pentomino board) and makes a decision. The observation is visual: the person *looks* at the objects. Same as in hearing words, *what* they see influences the information they get (same as what they hear). So, one question to ask is:

How does what they look at relate to what they finally choose?

To answer such a question, we need to do spatial analysis. As usual, we can start small and build up. One intuitive hypothesis, is that they would look at the object they selected more than they did on other objects. Hence our first question:

How much do people look at the selected object, compared to other objects?

#### STEP 4: Quantify how much people look at an object

We have to take into account:

* the position of the tile
* the position of the gaze pixel
* a tolerance area around the tile

NOTE: Tolerances are always relevant in any kind of analysis with real data. Think about the accuracy of the gaze tracking, noise in the data or simply random error factors.

We imagine a circle around the tile, with the radius of the circle being the tolerance. The larger the circle the greater the tolerance we allow. A gaze pixel will be considered "on the tile" if its *euclidean distance* from the center of the tile is smaller than the radius (in other words, if it is in the circle)

```python
#function that returns the board for an episode
def get_board(episodeno):
    episode_start = int(infos['episodes']['start_time'].ix[episodeno])
    episode_end = int(infos['episodes']['end_time'].ix[episodeno])
    boardslice = boards.ix[episode_start:episode_end]
    return boardslice.iloc[-1]
#function that returns the gaze points for an episode
def get_gaze(episodeno):
    episode_start = int(infos['episodes']['start_time'].ix[episodeno])
    episode_end = int(infos['episodes']['end_time'].ix[episodeno])
    return gazegood.ix[episode_start:episode_end]
```

```python
#function that gets the number of gaze points on an object
def get_number_of_gaze_points(gazepoints, tile, tolerance):
    count = 0
    for point in gazepoints['pixels']:
        if distance.euclidean((point.x, point.y), (tile.x, tile.y)) < tolerance:
            count += 1 
    return count
```

#### STEP 7: Get the rank of an object, based on how much it has been looked at

It would be interesting to *order* the objects by how much they have been looked at. We can do that easily by creating a dataframe as follows

```python
episodeno = 0
tolerance  = 75
aboard = pd.DataFrame(get_board(episodeno))
aboard['tile'] =  aboard[aboard.columns[0]]
del aboard[aboard.columns[0]]
aboard['looked_at'] = aboard['tile'].map(lambda x: get_number_of_gaze_points(get_gaze(episodeno), x, tolerance))
aboard['shape'] = aboard['tile'].map(lambda x: x.shape)
aboard['color'] = aboard['tile'].map(lambda x: x.color)
aboard['selected'] = aboard['tile'].map(lambda x: x.is_selected)
aboard['x'] = aboard['tile'].map(lambda x: x.x)
aboard['y'] = aboard['tile'].map(lambda x: x.y)
aboard
```

Now we simply sort the table based on the looked at points

```python
aboard.sort(columns='looked_at', ascending=False, inplace=True)
aboard
```

We can also re-index the table, then the index (+ 1) will simply be the rank

```python
aboard.reset_index(inplace=True)
aboard['rank'] =aboard.index + 1
aboard
```

#### STEP 8: Get the relative percentage looking at the selected piece

We would like to quantify how much they look at a piece compared to the other pieces.

To get the percentage of each piece, we simply add the points together to get the total, then do a simple division

```python
total = aboard['looked_at'].sum()
aboard['percentage'] = aboard['looked_at'].map(lambda x: round(100. * x / total, 2))
aboard
```

#### STEP 9: Get the rank amd percentage of the selected piece from all episodes

In order to get the rank of the single episode above, we simply *subset* for the selected pieces, of which there is always only one!

```python
rank = list(aboard['rank'][aboard['selected']])[0] #item 0 of the list is the one item in the list
percentage = list(aboard['percentage'][aboard['selected']])[0]
rank, percentage
```

So we simply need to loop over all epidodes

```python
tolerance  = 75
results = []
for episodeno in infos['episodes'].index:
    if episodeno in [36, 53, 142]:
        continue
    aboard = pd.DataFrame(get_board(episodeno))
    aboard['tile'] =  aboard[aboard.columns[0]]
    del aboard[aboard.columns[0]]
    aboard['looked_at'] = aboard['tile'].map(lambda x: get_number_of_gaze_points(get_gaze(episodeno), x, tolerance))
    aboard['selected'] = aboard['tile'].map(lambda x: x.is_selected)
    aboard.sort(columns='looked_at', ascending=False, inplace=True)
    aboard.reset_index(inplace=True)
    aboard['rank'] =aboard.index + 1
    total = aboard['looked_at'].sum()
    aboard['percentage'] = aboard['looked_at'].map(lambda x: round(100. * x / total, 2))
    rank = list(aboard['rank'][aboard['selected']])[0] #item 0 of the list is the one item in the list
    percentage = list(aboard['percentage'][aboard['selected']])[0]
    result = {'episodeno': episodeno, 'rank': rank, 'percentage': percentage}
    results.append(result)
RESULTS = pd.DataFrame(results)
RESULTS
```
